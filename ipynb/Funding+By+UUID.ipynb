```python
#creates a new dataframe that's empty
fund_df = pd.DataFrame() 
fund_df["uuid"] = ""
fund_df["Funding Type"] = ""
fund_df["Money Raised"] = ""
fund_df["Announced On"] = ""

index = 0

#org_json["data"]["items"][0 ]["properties"]["name"]
#print(len(org_json["data"]["paging"]))
#print(len(org_json["data"]["items"]))
#print(org_json["data"]["paging"])
#print(org_json["data"]["items"])
#print(len(org_json["data"]["items"]))

# Created my api key for Crunchbase
url = "https://api.crunchbase.com/v/3/funding-rounds/?user_key="
api_key = "8a3efb5b5136f95bff0593ebb94994e3"

# get 10 pages of org items
for x in range(1,11):
    # Build query URL
    query_url = url + api_key + "&page=" + str(x)
    # Get Crunchbase data
    funding_json = req.get(query_url).json()
    print(query_url)

    # Loop through a page of items 
    for i in range(0,len(funding_json["data"]["items"])):
        # Save the user class info into variables
        uuid  = funding_json["data"]["items"][i]["uuid"]
        fundingtype  = funding_json["data"]["items"][i]["properties"]["funding_type"]
        raised  = funding_json["data"]["items"][i]["properties"]["money_raised"]
        date = funding_json["data"]["items"][i]["properties"]["announced_on"]
        
               
        # Set the cell info for Real Name, Tweets, Followers, Following, Favorites Count
        fund_df.set_value(index, "uuid", uuid)
        fund_df.set_value(index, "Funding Type", fundingtype)
        fund_df.set_value(index, "Money Raised", raised)
        fund_df.set_value(index, "Announced On", date)
        index += 1

fund_df.sort_values("uuid")    
fund_df.head(10)
```
